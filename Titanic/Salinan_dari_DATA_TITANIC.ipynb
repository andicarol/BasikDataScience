#Dealing with Missing Value

```python
df_titanic.head(10)
```

Untuk kasus ini, kita bisa lihat nama customer pertama memiliki nilai Nan pada Columns Cabin


Langkah
*   Kita Bisa Mengganti Dengan nilai yang wajar
*   kita bisa membuang observasi yang mengandung missing value



sekarang kita check berapa banyak missing value dari setiap columns

```python
df_titanic.isna().sum(axis=0)
```

nah kita bisa melihat bahwa ada 3 columns (age, cabin dan embarked) yang memiliki missing value, sekarang kita handle untuk column **age**

```python
df_titanic['Age'].plot(kind='hist')
```

kita sudah melihat persebaran dari Column Age, sekarang kita cari rataan dari Column Age,

```python
rataan_umur = df_titanic['Age'].mean()
```

nah kalau sudah menemukan rataan umur, kita masukan saja nilai rataan umur ke dalam Missing Value yang ada di Column Age

```python
df_titanic['Age'] = df_titanic['Age'].fillna(rataan_umur)
```

```python
df_titanic.isna().sum(axis = 0)
```

kita bisa melihat sekarang missing values pada columns Age sudah menjadi 0, untuk cara ini bisa di gunakan untuk mencari model saja, apa bila ingin menganalisis atau mencari machine learning lebih baik menanyakan perihal data missing values kepada pihak Database operator yang bertanggung jawab

sekarang kita check untuk missing value pada Column Embark

```python
df_titanic[df_titanic['Embarked'].isna()]
```

nah karena datanya sedikit banget cuman 2, jadi kita bisa lihat charakteristik datanya seperti apa

karena data tidak terlalu berpengaruh (sedikit) jadi kita bisa hilangkan saja atau eleminasi

```python
df_titanic = df_titanic[-df_titanic['Embarked'].isna()]
```

#How to Overcome the outlier

## Numeric Outlier

```python
df_titanic.describe()
```

di sini bisa kita lihat untuk max column fare terjadi gap value ada yang tarif 0 dan ada yang tarif paling tinggi 512

```python
df_titanic.sort_values('Fare')
```

untuk asumsi saya fare yang 0 ini kemungkinan adalah outlier atau nahkoda kapal jadi gak bayar, tapi saya coba mengisi Column Fare yang kosong ini berdsarkan Pclass

```python
df_titanic.groupby('Pclass').agg({'Fare' : 'mean'})
```

hati - hati terkecoh, jangan langsung pake hasil yang di atas, jadi kita harus menghilngkan Fare yang bernilai 0,

```python
bool_titanic = df_titanic['Fare'] > 0
df_titanic[bool_titanic].groupby('Pclass').agg({'Fare' : 'mean'})
```

jadi kita harus mengkondisikan column Fare lebih dari 0
kemudian masukan rumus yang tadi


kemudian kita pake syntax loc untuk merubah value 0 menjadi value yang baru saja kita dapatkan

```python
df_titanic.loc[(df_titanic['Pclass']==1) & (df_titanic['Fare']==0),'Fare' ] = 86.207715
df_titanic.loc[(df_titanic['Pclass']==2) & (df_titanic['Fare']==0),'Fare' ] = 21.358661
df_titanic.loc[(df_titanic['Pclass']==3) & (df_titanic['Fare']==0),'Fare' ] = 13.787875
```

```python
df_titanic.sort_values('Fare')
```

#String Processing


Langkah
*   Mengekstrak informasi tertentu dari suatu string
*   memperbaiki data yang tidak konsistens



Pada kasus titanic tidak terdapat masalah pada string outlier, jadi saya membuat contoh untuk menggunakan syntax string processing

```python
df_wrong = pd.DataFrame(['Q','S','C','c','Q','S'],columns=['embarked'])
```

```python
df_wrong['embarked'].value_counts()
```

bisa kita liat ada outlier kategorikal, dimana ada c sebagai outlier dari embarked, dan harus kita handle

```python
df_wrong['embarked'] = df_wrong['embarked'].replace('c','C')
```

```python
df_wrong
```

```python
df_titanic['Title'] = df_titanic['Name'].str.split(', ').str[1].str.split('. ').str[0]
```

```python
df_titanic['Title'].value_counts()
```

dari sini kita bisa melihat tittle dari setiap naman, kemudian saya ingin memisahkan gelar biasa dan gelar unik berdsarkan persebarannya

```python
df_titanic['Title_Category'] = df_titanic['Title'].isin(['Mr','Miss','Mrs']).map({True : 'Normal_Title', False : 'Unique_Title'})
```

```python
df_titanic['Title_Category'] = df_titanic['Title_Category'].map({'Normal_Title' : 0, 'Unique_Title' : 1})
```

###Mean Encoding

digunakan untuk variable categori banyak dan tidak berururtan contohnya column cabin
